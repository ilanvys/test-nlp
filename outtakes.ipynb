```python
# Group by country to calculate the average sentiment score
sentiment_by_country = df.groupby('Country')['sentiment_score'].mean().reset_index()

# Create a choropleth map
fig_choropleth = px.choropleth(
    sentiment_by_country,
    locations='Country',
    locationmode='country names',
    color='sentiment_score',
    hover_name='Country',
    color_continuous_scale=px.colors.sequential.Plasma,  # Use a visually appealing color scale
    title='Average Sentiment Score by Country',
    labels={'sentiment_score': 'Average Sentiment Score'},
    range_color=(-3, 3)  # Adjust based on your sentiment score range
)

# Update layout for better aesthetics
fig_choropleth.update_layout(
    title_font_size=24,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular',  # Use a different map projection for aesthetics
        showland=True,
        landcolor="white",
        lakecolor="LightBlue"
    ),
    coloraxis_colorbar=dict(
        title="Sentiment Score",
        tickvals=[-3, -2, -1, 0, 1, 2],
        ticktext=['Extremely Negative', 'Negative', 'Slightly Negative', 'Neutral', 'Slightly Positive', 'Positive'],
    ),
    plot_bgcolor='rgba(0,0,0,0)'
)

fig_choropleth.show()
```

```python
# Function to convert country names to ISO 3-letter codes
def country_to_iso3(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except LookupError:
        return None

# Convert country names to ISO 3-letter codes
sentiment_by_country['iso_alpha'] = sentiment_by_country['Country'].apply(country_to_iso3)

# Drop any rows where conversion failed
sentiment_by_country = sentiment_by_country.dropna(subset=['iso_alpha'])

# Create a choropleth map using ISO 3-letter country codes
fig_choropleth = px.choropleth(
    sentiment_by_country,
    locations='iso_alpha',
    color='sentiment_score',
    hover_name='Country',
    color_continuous_scale=px.colors.sequential.Plasma,  # Use a visually appealing color scale
    title='Average Sentiment Score by Country',
    labels={'sentiment_score': 'Average Sentiment Score'},
    range_color=(-3, 3)  # Adjust based on your sentiment score range
)

# Update layout for better aesthetics
fig_choropleth.update_layout(
    title_font_size=24,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular',  # Use a different map projection for aesthetics
        showland=True,
        landcolor="white",
        lakecolor="LightBlue"
    ),
    coloraxis_colorbar=dict(
        title="Sentiment Score",
        tickvals=[-3, -2, -1, 0, 1, 2],
        ticktext=['Extremely Negative', 'Negative', 'Slightly Negative', 'Neutral', 'Slightly Positive', 'Positive'],
    ),
    plot_bgcolor='rgba(0,0,0,0)'
)

fig_choropleth.show()
```
